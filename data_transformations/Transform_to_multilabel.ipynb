# Transform Data Security

```python
# get unique labels
import numpy as np
import pandas as pd
dataframe = pd.read_csv("Data_Security.csv")
value_counts = dataframe['Security Measure'].value_counts()
print(value_counts)
```

```python
# import numpy as np
# import pandas as pd
# dataframe = pd.read_csv("Data_Security.csv")
dataframe.drop('category', axis=1, inplace=True)
dataframe = dataframe[~dataframe.duplicated(subset=['segment', 'Security Measure'])]
print((dataframe['Security Measure'].iloc[0]))


def label_to_vector(label, labels, count):
    """

    Returns a vector representing the label passed as an input.

    Args:
        label: string, label that we want to transform into a vector.
        labels: dictionary, dictionary with the labels as the keys and indexes as the values.
    Returns:
        vector: np.array, 1-D array of lenght 12.

    """

    vector = np.zeros((count), dtype=np.int64)
    try:

        index = labels[label]

        vector[index] = 1

    except KeyError:

        vector = np.zeros((count), dtype=np.int64)

    return vector

labels = {'Generic': 0,'Data access limitation': 1,'Secure data transfer': 2,'Other': 3,
          'Privacy/Security program': 4, 'Secure data storage': 5, 'Privacy review/audit': 6,
          'Secure user authentication': 7, 'Unspecified': 8, 'Privacy training': 9}


dataframe['Security Measure'] = dataframe['Security Measure'].apply(lambda x: label_to_vector(x, labels, 10)) # returns one hot encoding in a 1D-12D vector

labels_data = dataframe[[ 'segment', 'Security Measure']]
labels = labels_data.groupby("segment").sum() # since segments can have many labels, it sums label vectors of the same segment, resuling in a vector with all the labels of the segment marked as 1, other values as 0 (summing together one hot encodings)

labels = labels.reset_index()
labels.head() # reindex this
labels.shape
# print(labels.columns)
labels['Security Measure'] = labels['Security Measure'].apply(lambda x: x.tolist()) # convert to list to be able to work with it later


labels.to_csv("Data_Security2.csv", index=False)

labels.head()
#print(type(labels['Security Measure'][0]))
```
